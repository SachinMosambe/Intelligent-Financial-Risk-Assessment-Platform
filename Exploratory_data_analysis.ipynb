```python
scenario_dist = df.groupby("emi_scenario")["emi_eligibility"].value_counts(normalize = True).unstack().fillna(0)

plt.figure(figsize = (12,5))
sns.barplot(x = scenario_dist.index, y = scenario_dist['Eligible'], color = 'purple')
plt.title("Approval Rate by EMI Scenario")
plt.xlabel("Approval Rate")
plt.ylabel("Approval rate")
plt.show()

```

```python
scenario_dist = df.groupby("emi_scenario")["emi_eligibility"].value_counts(normalize=True).unstack().fillna(0)

scenario_dist.plot(kind="bar", stacked=True, figsize=(10,6), colormap="Set2")
plt.title("EMI Eligibility Distribution Across Scenarios")
plt.ylabel("Proportion")
plt.xlabel("EMI Scenario")
plt.legend(title="EMI Eligibility")
plt.show()
```

```python
gender_dist = (
    df.groupby("gender")["emi_eligibility"]
      .value_counts(normalize=True)
      .unstack()
      .fillna(0)
)

gender_dist.plot(kind="bar", stacked=True, figsize=(6,4), colormap="Accent")
plt.title("EMI Eligibility by Gender")
plt.ylabel("Proportion")
plt.show()
```

```python
edu_dist = (
    df.groupby("education")["emi_eligibility"]
      .value_counts(normalize=True)
      .unstack()
      .fillna(0)
)

edu_dist.plot(kind="bar", stacked=True, figsize=(8,5), colormap="Paired")
plt.title("EMI Eligibility by Education")
plt.ylabel("Proportion")
plt.xticks(rotation=45)
plt.show()
```

```python
emp_dist = (
    df.groupby("employment_type")["emi_eligibility"]
      .value_counts(normalize=True)
      .unstack()
      .fillna(0)
)

emp_dist.plot(kind="bar", stacked=True, figsize=(8,5), colormap="Dark2")
plt.title("EMI Eligibility by Employment Type")
plt.ylabel("Proportion")
plt.xticks(rotation=30)
plt.show()
```

```python
num_cols = ["monthly_salary","years_of_employment","credit_score","bank_balance"]

summary = df.groupby("emi_eligibility")[num_cols].agg(["mean","median"])
print(summary)
```

```python

df["debt_to_income"] = df["current_emi_amount"] / df["monthly_salary"].replace(0, np.nan)
                           
df["expense_to_income"] = (df["monthly_rent"] + df["school_fees"] + df["college_fees"] +
                           df["travel_expenses"] + df["groceries_utilities"] + 
                           df["other_monthly_expenses"]) / df["monthly_salary"].replace(0, np.nan)

```

```python
# Employment stability (years in job + company_type weight)
df["employment_stability_score"] = df["years_of_employment"] * df["employment_type"].map({
    "Permanent": 1.2,
    "Contract": 0.8,
    "Self-Employed": 1.0
}).fillna(1.0)

# Credit risk score buckets
df["credit_risk_score"] = pd.cut(
    df["credit_score"],
    bins=[0, 600, 700, 800, 900],
    labels=["High Risk", "Moderate Risk", "Good", "Excellent"]
)

# Savings buffer = (Bank balance + Emergency fund) / Monthly salary
df["savings_buffer_ratio"] = (df["bank_balance"] + df["emergency_fund"]) / df["monthly_salary"].replace(0, np.nan)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder,LabelEncoder
```

```python
le = LabelEncoder()
df["emi_eligibility"] = le.fit_transform(df["emi_eligibility"])
df["existing_loans"] = le.fit_transform(df["existing_loans"])
```

```python

nominal_col = ['gender','marital_status','employment_type','company_type','emi_scenario','house_type']

ohe = OneHotEncoder(drop ='first', handle_unknown='ignore')

encoded_array = ohe.fit_transform(df[nominal_col]).toarray()

encoded_df = pd.DataFrame(encoded_array, columns = ohe.get_feature_names_out(nominal_col))
encoded_df = encoded_df.astype(int)

df = pd.concat([df.drop(columns = nominal_col),encoded_df], axis = 1)
```

```python
from sklearn.preprocessing import OrdinalEncoder

education_order = ['Professional', 'Graduate', 'High School', 'Post Graduate']
risk_order = ['Moderate Risk', 'Good', 'High Risk', 'Excellent']

# Select multiple columns
ordinal_cols = ['education', 'credit_risk_score']

# Create encoder with category lists (order matters)
ordinal_encoder = OrdinalEncoder(categories=[education_order, risk_order])

# Fit + transform
df[ordinal_cols] = ordinal_encoder.fit_transform(df[ordinal_cols]).astype(int)
```

```python
columns = ['monthly_salary', 'years_of_employment', 'monthly_rent', 'family_size',
       'dependents', 'school_fees', 'college_fees', 'travel_expenses',
       'groceries_utilities', 'other_monthly_expenses', 'current_emi_amount',
       'credit_score', 'bank_balance', 'emergency_fund', 'requested_amount',
       'requested_tenure', 'max_monthly_emi']
skewness = df[columns].skew()
skewness
```

```python
high_skew_features = skewness[abs(skewness)>1].index.tolist()
high_skew_features
```

```python
positive_skew_features = skewness[(skewness>0.5) & (skewness<1.0)].index.tolist()
negetive_skew_features = skewness[(skewness<-0.5) & (skewness>-1.0)].index.tolist()
print(positive_skew_features)
print(negetive_skew_features)
```

```python
df[positive_skew_features] = df[positive_skew_features].apply(lambda x:np.log1p(x))
```

```python
df[high_skew_features] = df[high_skew_features].apply(lambda x:np.log1p(x))
```

```python
less_skewed_features = skewness[abs(skewness)<0.5].index.tolist()
less_skewed_features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df[less_skewed_features] = scaler.fit_transform(df[less_skewed_features])
```
